## Lab 1) 매출 파일 처리

```python
def process_sales(input_file="sales.txt", output_file="summary.txt"):
    try:
        with open(input_file, "r", encoding="utf-8") as infile:
            sales = [int(line.strip()) for line in infile if line.strip().isdigit()]
        total = sum(sales)
        avg = total / len(sales)
        with open(output_file, "w", encoding="utf-8") as outfile:
            outfile.write(f"총매출 = {total}\n")
            outfile.write(f"평균 일매출 = {avg}\n")
        print(f"총매출 = {total}")
        print(f"평균 일매출 = {avg}")
    except FileNotFoundError:
        print(f"입력 파일 {input_file} 을(를) 찾을 수 없습니다.")
    except Exception as e:
        print("오류 발생:", e)
```

## Lab 2) 각 문자 횟수 세기

```python
def count_characters(filename="input.txt"):
    try:
        counts = {}
        with open(filename, "r", encoding="utf-8") as infile:
            for line in infile:
                for ch in line:
                    counts[ch] = counts.get(ch, 0) + 1
        print(counts)
        return counts
    except FileNotFoundError:
        print(f"파일 {filename} 을(를) 찾을 수 없습니다.")
    except Exception as e:
        print("오류 발생:", e)
```

## Lab 3) 디렉토리 안의 파일 처리

```python
import os

def search_python_lines(directory="."):
    try:
        for filename in os.listdir(directory):
            if os.path.isfile(filename):
                with open(filename, "r", encoding="utf-8", errors="ignore") as f:
                    for line in f:
                        if "Python" in line:
                            print(f"{filename}: {line.strip()}")
    except Exception as e:
        print("오류 발생:", e)
```
