```python
## Lab 1) 넘파이 배열 생성하기
import numpy as np

# 1부터 9까지 1차원 배열
a = np.arange(1, 10)
print("a =", a)

# 3x3 행렬로 변환
b = a.reshape(3, 3)
print("b =\n", b)

# 0으로 가득 찬 배열
c = np.zeros((2, 4))
print("c =\n", c)

# 1로 가득 찬 배열
d = np.ones((3, 3))
print("d =\n", d)

# 난수 배열 생성
e = np.random.rand(2, 3)
print("e =\n", e)
```

```python
## Lab 2) 넘파이 배열의 슬라이싱과 인덱싱
import numpy as np

arr = np.arange(1, 21).reshape(4, 5)
print("arr =\n", arr)

# 특정 원소
print("arr[1, 2] =", arr[1, 2])

# 첫 번째 행
print("첫 번째 행 =", arr[0, :])

# 두 번째 열
print("두 번째 열 =", arr[:, 1])

# 가운데 2×3 부분 배열
print("가운데 2x3 =\n", arr[1:3, 1:4])

# 마지막 두 행
print("마지막 두 행 =\n", arr[2:, :])

# 짝수 위치 요소만 추출
print("짝수 값만 =", arr[arr % 2 == 0])
```

```python
## Lab 3) 벡터 연산 실습
import numpy as np

v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

# 벡터 덧셈
print("v1 + v2 =", v1 + v2)

# 벡터 내적(dot product)
print("내적 =", np.dot(v1, v2))

# 스칼라 곱
print("3 * v1 =", 3 * v1)

# 벡터 길이(norm)
print("v1의 길이 =", np.linalg.norm(v1))
print("v2의 길이 =", np.linalg.norm(v2))
```

```python
## Lab 4) 행렬 연산 실습
import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print("A =\n", A)
print("B =\n", B)

# 행렬 덧셈
print("A + B =\n", A + B)

# 행렬 곱(dot)
print("A dot B =\n", A.dot(B))

# 전치행렬
print("A^T =\n", A.T)

# 역행렬
print("A의 역행렬 =\n", np.linalg.inv(A))

# 행렬식
print("det(A) =", np.linalg.det(A))
```
