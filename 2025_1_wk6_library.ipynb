배열의 인덱싱이나 슬라이싱을 할 때 주의할 점은 차원과 0부터 시작한다는 점, 오른쪽 끝은 포함되지 않는 점 등이다.





```python
a = np.arange(1, 10)
print(a)
```

```python
#슬라이싱과 인덱싱은 본질적으로 차이가 있다. 똑같은 6을 슬라이싱하면 다음과 같이 될 것이다.
print(b[1:2, 2:])
```

```python
#  4 5 6 을 슬라이싱할 때도 다음 두 가지는 차이가 있다.
print(b[1, :])
print(b[1:2, :])
```

```python
arr = np.array([[10, 20, 30], [40, 50, 60], [70, 80, 90]])

# 특정 원소 접근
print(arr[0, 2])  # 30

# 슬라이싱
print(arr[:, 1])  # [20 50 80] (두 번째 열)
print(arr[1:, :2])  # [[40 50] [70 80]]

# Boolean 인덱싱
print(arr[arr > 50])  # [60 70 80 90]

# Fancy 인덱싱
indices = [0, 2]
print(arr[indices, :])  # 첫 번째와 세 번째 행 가져오기
```

# 6.1.3 배열의 연산

배열이 리스트와 비교했을 때 좋은 점은 연산을 가지고 있다는 것이다. 배열에서 덧셈, 뺄셈 등은 각 성분별 덧셈, 뺄셈으로 쉽게 정의된다. 행렬의 곱셈은 이전에 이야기한 것처럼 dot 함수로 표현한다. 몇 가지 연산을 알아보기로 하자.

만약 두 행렬이 다음과 같이 정의되어 있을 때 행렬의 곱은 어떻게 계산할까?

$$
\begin{align}
x =
\begin{bmatrix}
1 &  0 & 1 \\
0  & 2 & 1 \\
\end{bmatrix}
\end{align}
$$

$$
\begin{align}
y =
\begin{bmatrix}
-1 & 1 \\
0  & 2  \\
1 &  -1\\
\end{bmatrix}
\end{align}
$$

```python
x = np.array([[1, 0, 1], [0, 2, 1]])
y= np.array([[-1, 1], [0, 2], [1, -1]])
print(x)
print(y)
```

```python
#행렬의 곱  xy는 다음과 같이 계산한다.
print(np.dot(x,y))
x.dot(y)
```

# 실습과제 6-4

4차함수 $f(x)=x(x+1)(x-1)(x-2)$ 과 $g(x)=x(x+1)(x+2)(x-1)$의 그래프를 그려보아라. 그리드와 범례도도 넣어보시오.

```python
def f_3(x, w):
    return (x-1)*x*(x+1)*(x-w)

# 아래 코드를 적당히 고쳐 보세요.
x=np.linspace( -3, 3, 100   ) # -3에서 3까지 범위 정하기
plt.plot(x, f_3(x, 2)  , color='black', label ='$w=2$' )
plt.plot(x, f_3(x,-2)   ,color='cornflowerblue', label= '$w=-2$')
plt.legend(loc='best')
plt.ylim( -5 , 100 )
plt.title('$f_3(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid(True)
plt.show()
```

# 실습과제 6-5-1(제출하지 마시오.)

아래와 같이 주어진 함수가 있을 때, 이를 그래프로 표현해 보시오.

함수는 $ y= x^3 + 3x^2 +2$ 그래프의 제목은 'graph of polynomial of degree 3' 라고 하고 x축, y축 이름은 x, y로 한다. x의 범위는 -10에서 10까지 그래프의 색은 빨간색, 선 모양은 dashed (점선) 로 한다. legend 는 'order 3' 라고 넣는다.

```python
import matplotlib.pyplot as plt
def f_4(x):
    return x**3 +3*x**2 +2
x=np.linspace( -3, 3, 100   ) # -3에서 3까지 범위 정하기
plt.plot(x, f_4(x)  , 'r--', label ='order 3' )
plt.xlabel('x')
# Set the y axis label of the current axis.
plt.ylabel('y')
# Set a title
plt.title('graph of polynomial of degree 3')
plt.legend(loc='best')
#여기에 알맞은 코드를 넣으시오.
plt.show()
```

$\bullet$ 데이터 프레임의 다양한 인덱싱 방법

- df[val] :	Select single column or sequence of columns from the DataFrame

- df.loc[val]	: Selects single row or subset of rows from the DataFrame by label

- df.loc[:, val]	: Selects single column or subset of columns by label

- df.loc[val1, val2]	: Select both rows and columns by label

- df.iloc[where]	: Selects single row or subset of rows from the DataFrame by integer position

- df.iloc[:, where]	: Selects single column or subset of columns by integer position

- df.iloc[where_i, where_j] : Select both rows and columns by integer position

위에서 df.loc 과 df.iloc의 차이를 알아두면 좋을 것 같다. df.loc[] 은 행이름과 같이 레이블이름으로 접근하는 것이고 df.iloc[]은 0, 1, 2,...처럼 array 인덱싱처럼 접근하는 것이다.
