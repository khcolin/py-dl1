## Lab 1) TV 클래스 정의

```python
class Television:
    def __init__(self, channel: int, volume: int, on: bool):
        self.channel = channel
        self.volume = volume
        self.on = on

    def show(self):
        print(self.channel, self.volume, self.on)

    def setChannel(self, ch: int):
        self.channel = ch

    def getChannel(self) -> int:
        return self.channel

t = Television(9, 10, True)
t.show()
t.setChannel(11)
t.show()
```

## Lab 2) 원(Circle) 클래스 정의

```python
import math

class Circle:
    def __init__(self, radius: float):
        self.radius = radius

    def getArea(self) -> float:
        return math.pi * (self.radius ** 2)

    def getPerimeter(self) -> float:
        return 2 * math.pi * self.radius

c = Circle(10)
print("원의 면적", c.getArea())
print("원의 둘레", c.getPerimeter())
```

## Lab 3) 자동차(Car) 클래스 정의

```python
class Car:
    def __init__(self, speed: int, color: str, model: str):
        self.speed = speed
        self.color = color
        self.model = model

    def drive(self):
        self.speed = 60

myCar = Car(0, "blue", "E-class")
print("자동차 객체를 생성하였습니다.")
print("자동차의 속도는", myCar.speed)
print("자동차의 색상은", myCar.color)
print("자동차의 모델은", myCar.model)
myCar.drive()
print("자동차의 속도는", myCar.speed)
```

## Lab 4) 은행 계좌(Account) 클래스

```python
class Account:
    def __init__(self, init_balance: int = 0):
        self.__balance = init_balance

    def deposit(self, amount: int):
        if amount < 0:
            raise ValueError("입금 금액은 0 이상이어야 함")
        self.__balance += amount
        print(f"통장에 {amount} 가 입금되었음")

    def withdraw(self, amount: int):
        if amount < 0:
            raise ValueError("출금 금액은 0 이상이어야 함")
        if amount > self.__balance:
            raise ValueError("잔액 부족")
        self.__balance -= amount
        print(f"통장에서 {amount} 가 출금되었음")

    def getBalance(self) -> int:
        return self.__balance

acc = Account(200)
acc.withdraw(100)
acc.deposit(10)
```

## Lab 6) 2차원 벡터(Vector) 연산

```python
class Vector2D:
    def __init__(self, a: float, b: float):
        self.a = a
        self.b = b

    def __add__(self, other):
        return Vector2D(self.a + other.a, self.b + other.b)

    def __sub__(self, other):
        return Vector2D(self.a - other.a, self.b - other.b)

    def __str__(self):
        return f"({self.a}, {self.b})"

v1 = Vector2D(0, 1)
v2 = Vector2D(1, 0)
print(f"{v1} + {v2} = {v1 + v2}")
```

## Lab 7) Book 클래스 연산자 오버로딩

```python
class Book:
    def __init__(self, title: str = "", pages: int = 0):
        self.title = title
        self.pages = pages

    def __str__(self):
        return self.title

    def __gt__(self, other) -> bool:
        return self.pages > other.pages

    def __add__(self, other) -> int:
        return self.pages + other.pages

book1 = Book("Magic of Python", 600)
book2 = Book("Master of Python", 700)
print(book1 > book2)
print(book1 + book2)
```
