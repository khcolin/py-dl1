```python
## Lab 1) 내장 함수 예제

invitations = ["Kim", "Lee", "Park", "Choi"]
persons = [1, 3, 0, 6]


total = sum(persons)
print("총 인원 수:", total)


any_coming = any(persons)
print("한 사람이라도 오는가?:", any_coming)

all_coming = all(p > 0 for p in persons)
print("모든 그룹이 전부 오는가?:", all_coming)


max_group = max(persons)
print("가장 많이 오는 그룹 인원:", max_group)
```

```python
## Lab 2) 키를 이용한 정렬 예제

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __repr__(self):
        return f"<이름: {self.name}, 나이: {self.age}>"

people = [
    Person("홍길동", 20),
    Person("김철수", 35),
    Person("최자영", 38)
]


sorted_people = sorted(people, key=lambda p: p.age)
print(sorted_people)
```

```python
## Lab 3) 람다식으로 온도 변환하기

f_temp = [0, 10, 20, 30, 40, 50]

c_temp = map(lambda T: (5.0/9.0) * (T - 32.0), f_temp)

print(list(c_temp))
```

```python
## Lab 4) 람다식으로 데이터 처리하기

orders = [
    ["1", "재킷", 5, 120000],
    ["2", "셔츠", 6, 24000],
    ["3", "바지", 3, 50000],
    ["4", "코트", 6, 300000]
]


result = list(map(lambda order: (order[0], order[2] * order[3]), orders))
print(result)

```

```python
## Lab 5) 동전 던지기 게임

import random

while True:
    answer = input("동전 던지기를 계속하시겠습니까?( yes, no) ").strip().lower()
    if answer == "yes":
        coin = random.choice(["head", "tail"])
        print(coin)
    elif answer == "no":
        break
    else:
        print("yes 또는 no 로 입력하세요.")
```
