```python
# Lab 1: 문자열의 공통 문자
s1 = input("첫 번째 문자열: ")
s2 = input("두 번째 문자열: ")

common = (set(s1) & set(s2)) - {" "}
print("공통적인 글자:", " ".join(sorted(common)))
```

```python
# Lab 2: 중복되지 않은 단어의 개수 세기
import string
text = input("입력 텍스트: ").strip()
tbl = str.maketrans("", "", string.punctuation)
tokens = [w.translate(tbl).lower() for w in text.split() if w.translate(tbl)]
uniq = set(tokens)
print("사용된 단어의 개수=", len(uniq))
print(uniq)
```

```python
# Lab 3: 영한 사전
english_dict = {"one": "하나", "two": "둘", "three": "셋", "four": "넷", "five": "다섯"}
word = input("단어를 입력하시오: ").strip().lower()
meaning = english_dict.get(word)
if meaning:
    print(meaning)
else:
    print("사전에 없는 단어이다.")
```

```python
# Lab 4: 친구 관리 프로그램
contacts = {}

def add_contact():
    name = input("이름: ").strip()
    phone = input("전화번호: ").strip()
    if not name or not phone:
        print("이름/전화번호가 비어있다.")
        return
    contacts[name] = phone
    print("추가 완료")

def delete_contact():
    name = input("삭제할 이름: ").strip()
    if name in contacts:
        contacts.pop(name)
        print("삭제 완료")
    else:
        print("해당 이름이 없음")

def search_contact():
    name = input("검색할 이름: ").strip()
    if name in contacts:
        print(f"{name} 의 전화번호: {contacts[name]}")
    else:
        print("해당 이름이 없음")

def print_all():
    if not contacts:
        print("연락처가 비어있다.")
        return
    for name, phone in sorted(contacts.items()):
        print(f"{name} 의 전화번호: {phone}")

while True:
    print("\n1. 연락처 추가\n2. 연락처 삭제\n3. 연락처 검색\n4. 연락처 출력\n5. 종료")
    sel = input("메뉴 항목을 선택하시오: ").strip()
    if sel == "1":
        add_contact()
    elif sel == "2":
        delete_contact()
    elif sel == "3":
        search_contact()
    elif sel == "4":
        print_all()
    elif sel == "5":
        print("프로그램 종료")
        break
    else:
        print("올바른 번호를 선택하시오.")
```

```python
# Lab 5: 학생 성적 처리
score_dic = {"Kim": [99, 83, 95], "Lee": [68, 45, 78], "Choi": [25, 56, 69]}
for name, scores in score_dic.items():
    avg = sum(scores) / len(scores) if scores else 0.0
    print(f"{name} 의 평균성적= {avg}")
```

```python
# Lab 6: 단어 카운터 만들기
import string
from collections import Counter
text_data = input("텍스트를 입력하시오: ")
tbl = str.maketrans("", "", string.punctuation)
words = [w.translate(tbl).lower() for w in text_data.split() if w.translate(tbl)]
counter = Counter(words)
for w, c in sorted(counter.items()):
    print(f"{w} 의 등장횟수= {c}")
```

```python
# Lab 7: 회문 검사하기
import re
s = input("문자열을 입력하시오: ")
norm = re.sub(r"[^0-9A-Za-z가-힣]", "", s).lower()
if norm == norm[::-1]:
    print("회문입니다.")
else:
    print("회문이 아닙니다.")
```

```python
# Lab 8: 머리 글자어 만들기
sentence = input("문자열을 입력하시오: ").strip()
acronym = "".join(word[0].upper() for word in sentence.split() if word)
print(acronym)
```

```python
# Lab 9: 이메일 주소 분석
email = input("이메일 주소를 입력하시오: ").strip()
if "@" not in email or email.count("@") != 1:
    print("올바른 이메일 형식이 아니다.")
else:
    user, domain = email.split("@", 1)
    print(email)
    print("아이디:", user)
    print("도메인:", domain)
```

```python
# Lab 10: 문자열 분석
s = input("문자열을 입력하시오: ")
result = {"digits": sum(ch.isdigit() for ch in s),
          "spaces": sum(ch.isspace() for ch in s),
          "alphas": sum(ch.isalpha() for ch in s)}
print(result)
```
